```python
available_balance=100000
withdraw_amt=float(input("Enter the Amount:"))
if(withdraw_amt < 100):
    print("Ensure to maintain enough balance")
elif(withdraw_amt% 100==0):
    balance=available_balance-withdraw_amt
    print("Withdraw Successfully")
    print("Remaining balance amount is:",balance)
else:
    print("Enter the right amount")
```
