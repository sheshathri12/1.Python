```python
num = int(input("Enter a number: "))
if num <= 1:
    print("Not a Prime Number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
```

```python
n=int(input("Enter the factorial number:"))
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
print(factorial(n))
```

```python
num=int(input("Enter the power number:"))
pow=len(str(num))
total=0
for d in str(num):
    total += int(d)**pow
if total == num:
    print("Amstrong Number")
else:
    print("Not an Amstrong Number")
```

```python
num = int(input("Enter decimal number: "))
binary = ""

while num > 0:
    remainder = num % 2
    binary = str(remainder) + binary
    num //= 2
    
print("Binary:", binary)
```
