```python
#5.Write a Python program to check if a number is prime. Expected Output (for input 7):
#7 is a prime number
n = int(input("Enter a number: "))
if n <= 1:
    print(n,"is not a Prime Number")
else:
    for i in range(2,n):
        if n % i == 0:
            print(n,"is not a Prime Number")
            break
    else:
       print(n,"is a Prime Number")

```
